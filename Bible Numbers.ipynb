# Numbers

The following section looks at various numbers presented in the book of Numbers.

```python
tribes = pd.DataFrame(
    data = [
        ['Reuben','1st census',46500],
        ['Simeon','1st census',59300],
        ['Gad','1st census',45650],
        ['Judah','1st census',74600],
        ['Issachar','1st census',54400],
        ['Zebulun','1st census',57400],
        ['Ephraim','1st census',40500],
        ['Manasseh','1st census',32200],
        ['Benjamin','1st census',35400],
        ['Dan','1st census',62700],
        ['Asher','1st census',41500],
        ['Naphtali','1st census',53400],
    ],
    columns = ['tribe','event','value']
).sort_values(by='value')
tribes
```

```python
bp = sns.barplot(x='tribe', y='value', hue='event', data=tribes)
bp.set(xlabel='Tribe', ylabel='People')
plt.xticks(rotation=30)
fig = bp.get_figure()
fig.savefig('census_1.jpg') 
```
